````python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Select relevant features for clustering (exclude 'City' to avoid leakage)
features = ['Price_SAR', 'Rental_Yield_Percent', 'Occupancy_Rate', 'Affordability_Index', 
            'Absorption_Rate', 'Distance_to_Center_km', 'Urban_Density_Index', 'Price_per_m2']

X = data[features]

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# K-means clustering (5 clusters for 5 cities)
kmeans = KMeans(n_clusters=5, random_state=42, n_init=10)
data['Cluster'] = kmeans.fit_predict(X_scaled)

# Show cluster counts and compare with city distribution
print("Cluster counts:")
print(data['Cluster'].value_counts().sort_index())
print("\nCity counts:")
print(data['City'].value_counts())
````

````python
# Example: Monte Carlo simulation to estimate average property price after 1 year with random price changes

n_simulations = 1000
n_properties = len(data)
price_changes = np.random.normal(loc=0.05, scale=0.10, size=(n_simulations, n_properties))  # 5% mean, 10% std

# Simulate new prices for each property in each simulation
simulated_prices = data['Price_SAR'].values * (1 + price_changes)

# Calculate average price per simulation
avg_simulated_prices = simulated_prices.mean(axis=1)

# Plot distribution of simulated average prices
plt.figure(figsize=(8, 5))
sns.histplot(avg_simulated_prices, bins=30, kde=True)
plt.title('Monte Carlo Simulation: Average Property Price After 1 Year')
plt.xlabel('Average Simulated Price (SAR)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

print(f"Expected average price after 1 year: SAR {avg_simulated_prices.mean():,.0f}")
print(f"95% confidence interval: SAR {np.percentile(avg_simulated_prices, 2.5):,.0f} - SAR {np.percentile(avg_simulated_prices, 97.5):,.0f}")
````
